```python

import pandas as pd

# Load Titanic dataset
# (Make sure titanic.csv is in the same folder as this notebook)
df = pd.read_csv("titanic_dataset.csv")

# Show first few rows
df.head()
```


## üß† Step 1: Inspecting the Data

Functions to explore:
- df.head()
- df.tail()
- df.info()
- df.describe()
- df.shape
- df.columns


```python

# Muestra los 5 primeros objetos del CSV
df.head()

# Muestra los primeros n√∫meros de objetos que indicas entre parentesis, por ejemplo (10)
df.head(10)

# Muestra los 5 √∫ltimos objetos del CSV
df.tail()

# Muestra los √∫ltimos n√∫meros de objetos que indicas entre parentesis, por ejemplo (10)
df.tail(10)

# Resumen del DataFrame (tipos de datos, valores no nulos...)
df.info()

# Genera estad√≠sticas de las columnas num√©ricas.
df.describe()

# Las dimensiones del DataFrame: (n√∫mero_de_filas, n√∫mero_de_columnas)
df.shape

# Una lista de los nombres de todas las columnas
df.columns
```


## üîç Step 2: Selecting Data

Functions to explore:
- df["column"]
- df[["col1", "col2"]]
- df.loc[]
- df.iloc[]


```python

# Mostrar el head (5 primeros) de la/las columna/s que indiques.
df["Age"].head()
df[["Sex", "Age", "Survived"]].head()

# Muestra las filas con √≠ndice de etiqueta 0, 1 y 2, solo para la columna 'Name'
df.loc[0:2, ["Name"]]

# Muestra las filas en posici√≥n 0, 1 y 2, y las columnas en posici√≥n 1, 2 y 3.
df.iloc[0:3, 1:4]
```


## üîé Step 3: Filtering Rows

Functions to explore:
- df[df["Age"] > 30]
- df.query("Sex == 'female' and Survived == 1")


```python

# Example: Filtering data
df[df["Age"] > 50].head()
df.query("Sex == 'female' and Survived == 1").head()

# Ense√±a los 5 primeros pasajeros que SOBREVIVIERON (Survived = 1)
df[df["Survived"] == 1].head()

# Ense√±a solo a los pasajeros que pagaron una TARIFA (Fare) superior a 247
df[df["Fare"] > 247].head()

# Ense√±a a los pasajeros de PRIMERA CLASE (Pclass = 1) que NO sobrevivieron (Survived = 0)
df[(df["Pclass"] == 1) & (df["Survived"] == 0)].head()


# Ense√±a a todas las MUJERES (Sex = 'female') que sobrevivieron (Survived = 1)
df.query("Sex == 'female' and Survived == 1").head()

# Ense√±a a los menores de 10 a√±os que viajaban en TERCERA CLASE (Pclass = 3)
df.query("Age < 10 and Pclass == 3").head()
```


## üßπ Step 4: Handling Missing Data

Functions to explore:
- df.isna()
- df.isna().sum()
- df.dropna()
- df.fillna()


```python

# Example: Check missing values
df.isna().sum()

# Fill missing ages with median
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Age"].head()

# Muestra un DataFrame booleano (True si es faltante, False si tiene valor)
df.isna().head()

# Cuenta el n√∫mero total de valores faltantes por columna
df.isna().sum()

# Elimina (borra) cualquier fila que contenga un valor que falte (como NaN o None).
df.dropna()

# Rellena los valores faltantes (NaN) en el DataFrame con el n√∫mero 0
df.fillna(0)

# Rellena los valores faltantes de la columna "Age" con el n√∫mero est√°tico 25
df["Age"].fillna(25)
```


## üìä Step 5: Grouping and Aggregating

Functions to explore:
- df.groupby("Sex")["Survived"].mean()
- df["Fare"].mean()
- df["Age"].median()


```python

# Example: Aggregation
df.groupby("Sex")["Survived"].mean()
df.groupby("Pclass")["Fare"].mean()

# Calcula la media (promedio) de la tarifa pagada por todos los pasajeros
df["Fare"].mean()

# Calcula la mediana  de la edad de todos los pasajeros
df["Age"].median()

# Muestra la tasa de supervivencia por g√©nero (la media).
df.groupby("Sex")["Survived"].mean()
```


## üìà Step 6: Sorting and Counting

Functions to explore:
- df.sort_values("Age")
- df["Sex"].unique()
- df["Pclass"].value_counts()


```python

# Example: Sorting and counting
df.sort_values("Age").head()
df["Pclass"].value_counts()

# Ordena el DataFrame de forma ascendente bas√°ndose en la columna "Age"
df.sort_values("Age").head()

# Ordena por la columna "Pclass" y, en caso de que sea la misma, ordena por "Age"
df.sort_values(["Pclass", "Age"]).head()

# Muestra una lista de todos los valores √öNICOS de la columna "Sex"
df["Sex"].unique()

# Muestra una lista de todos los valores √öNICOS de la columna "Embarked"
df["Embarked"].unique()

# Cuenta cu√°ntos pasajeros sobrevivieron (Survived=1) y cu√°ntos no (Survived=0)
df["Survived"].value_counts()
```


## ‚öôÔ∏è Step 7: Creating or Modifying Columns

Functions to explore:
- df.assign()
- df.apply()
- df["new_col"] = ...
- pd.concat()
- pd.merge()


```python

# Example: Create new column
df["Fare_per_Age"] = df["Fare"] / df["Age"]
df[["Age", "Fare", "Fare_per_Age"]].head()

# Crea un nuevo DataFrame con una columna nueva llamada 'Fare_x_2' multiplicando el Fare por dos.
df.assign(Fare_x_2=df["Fare"] * 2).head()

# .apply() para crear la columna 'Tamano_Familia' sumando 'SibSp' y 'Parch' m√°s 1 (para la propia persona)
# El 'axis=1' es para que la funci√≥n se aplique fila por fila.
df["Tamano_Familia"] = df.apply(lambda row: row["SibSp"] + row["Parch"] + 1, axis=1)
df[["SibSp", "Parch", "Tamano_Familia"]].head()
```


## üíæ Step 8: Exporting Data

Function to explore:

```python

# Example: Save cleaned data
df.to_csv("titanic_cleaned.csv", index=False)

# Es para guardar el csv actual en un archivo.
```
